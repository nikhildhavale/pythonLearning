```python
# HOUSE PRICES: EDA â†’ Preprocessing â†’ XGBoost Regression
# Required: pandas, numpy, seaborn, matplotlib, scikit-learn, xgboost, joblib

import os
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
import joblib
import warnings

from sklearn.model_selection import train_test_split, GridSearchCV, KFold
from sklearn.preprocessing import OneHotEncoder, StandardScaler
from sklearn.impute import SimpleImputer
from sklearn.pipeline import Pipeline
from sklearn.compose import ColumnTransformer
from sklearn.metrics import mean_squared_error, r2_score

import xgboost as xgb

sns.set(style="whitegrid")
plt.rcParams["figure.figsize"] = (10, 6)
warnings.filterwarnings("ignore")
RND = 42

# --------------------------------------------------
# 1. LOAD DATA
# --------------------------------------------------
assert os.path.exists("train_house.csv"), "Place House Prices 'train.csv' in the current directory."

df = pd.read_csv("train_house.csv")
print("Loaded train.csv: ", df.shape)
display(df.head())

# --------------------------------------------------
# 2. QUICK EDA
# --------------------------------------------------
print(df.info())

missing = df.isnull().sum().sort_values(ascending=False)
print("\nTop missing values:")
display(missing.head(20))

# Target distribution
plt.figure()
sns.histplot(df["SalePrice"], bins=40, kde=True)
plt.title("SalePrice Distribution")
plt.show()

print("SalePrice skew:", df["SalePrice"].skew())

# --------------------------------------------------
# 3. OUTLIER REMOVAL
# Common recommendation: remove extreme GrLivArea points
# --------------------------------------------------
outliers = df[(df["GrLivArea"] > 4000) & (df["SalePrice"] < 300000)].index
print("Dropping outliers:", list(outliers))

df = df.drop(index=outliers).reset_index(drop=True)
print("New shape:", df.shape)

# --------------------------------------------------
# 4. FEATURE ENGINEERING
# --------------------------------------------------
df["TotalBath"] = (
    df["FullBath"].fillna(0)
    + df["HalfBath"].fillna(0) * 0.5
    + df["BsmtFullBath"].fillna(0)
    + df["BsmtHalfBath"].fillna(0) * 0.5
)

df["TotalSF"] = (
    df["TotalBsmtSF"].fillna(0)
    + df["1stFlrSF"].fillna(0)
    + df["2ndFlrSF"].fillna(0)
)

df["MSSubClass"] = df["MSSubClass"].astype(str)

# Log-transform target to fix skew
df["SalePrice_log"] = np.log1p(df["SalePrice"])

# --------------------------------------------------
# 5. DEFINE FEATURES AND TARGET
# --------------------------------------------------
drop_cols = ["Id", "SalePrice", "SalePrice_log"]
predictor_cols = [c for c in df.columns if c not in drop_cols]
target_col = "SalePrice_log"

# Separate numeric & categorical
num_cols = df[predictor_cols].select_dtypes(include=[np.number]).columns.tolist()
cat_cols = [c for c in predictor_cols if c not in num_cols]

print("Numeric:", len(num_cols), "Categorical:", len(cat_cols))

# --------------------------------------------------
# 6. PREPROCESSING PIPELINE
# --------------------------------------------------

num_imputer = SimpleImputer(strategy="median")
num_scaler = StandardScaler()

cat_imputer = SimpleImputer(strategy="most_frequent")
ohe = OneHotEncoder(handle_unknown="ignore", sparse_output=False)

preprocessor = ColumnTransformer([
    ("num", Pipeline([("impute", num_imputer), ("scale", num_scaler)]), num_cols),
    ("cat", Pipeline([("impute", cat_imputer), ("ohe", ohe)]), cat_cols),
], remainder="drop")

# --------------------------------------------------
# 7. TRAIN-TEST SPLIT
# --------------------------------------------------
X = df[predictor_cols]
y = df[target_col].values  # log target

X_train, X_test, y_train, y_test = train_test_split(
    X, y, test_size=0.20, random_state=RND
)

print("Train shape:", X_train.shape, "Test shape:", X_test.shape)

# --------------------------------------------------
# 8. XGBOOST MODEL + PIPELINE
# --------------------------------------------------
xgb_model = xgb.XGBRegressor(
    objective="reg:squarederror",
    random_state=RND,
    n_estimators=200,
    learning_rate=0.1,
    max_depth=6,
    subsample=0.9,
    colsample_bytree=0.9,
    tree_method="auto"
)

pipeline = Pipeline([
    ("preproc", preprocessor),
    ("xgb", xgb_model)
])

print("Training baseline XGBoost model...")
pipeline.fit(X_train, y_train)
print("Done.")

# --------------------------------------------------
# 9. EVALUATE BASELINE MODEL
# --------------------------------------------------
y_pred_log = pipeline.predict(X_test)

rmse_log = np.sqrt(mean_squared_error(y_test, y_pred_log))
r2 = r2_score(y_test, y_pred_log)

print("\nBaseline RMSE (log scale):", rmse_log)
print("Baseline R2:", r2)

# Convert back to original scale
y_test_orig = np.expm1(y_test)
y_pred_orig = np.expm1(y_pred_log)
rmse_orig = np.sqrt(mean_squared_error(y_test_orig, y_pred_orig))

print("Baseline RMSE (Original Scale):", rmse_orig)

# --------------------------------------------------
# 10. RESIDUAL PLOT
# --------------------------------------------------
plt.figure()
plt.scatter(y_pred_log, y_test - y_pred_log, alpha=0.5)
plt.axhline(0, color="black", linestyle="--")
plt.title("Residual Plot (log scale)")
plt.xlabel("Predicted log price")
plt.ylabel("Residual")
plt.show()

# --------------------------------------------------
# 11. FEATURE IMPORTANCE
# --------------------------------------------------
ohe_cols = pipeline.named_steps["preproc"].named_transformers_["cat"].named_steps["ohe"].get_feature_names_out(cat_cols)
feature_names = num_cols + list(ohe_cols)

model_fitted = pipeline.named_steps["xgb"]
importances = model_fitted.feature_importances_

feat_df = pd.DataFrame({"feature": feature_names, "importance": importances})
feat_df = feat_df.sort_values("importance", ascending=False)

print("\nTOP 20 FEATURES:")
display(feat_df.head(20))

plt.figure(figsize=(8,10))
sns.barplot(x="importance", y="feature", data=feat_df.head(20))
plt.title("XGBoost Feature Importances")
plt.show()

# --------------------------------------------------
# 12. GRID SEARCH (SMALL)
# --------------------------------------------------
param_grid = {
    "xgb__n_estimators": [200, 400],
    "xgb__max_depth": [4, 6],
    "xgb__learning_rate": [0.05, 0.1],
}

kf = KFold(n_splits=3, shuffle=True, random_state=RND)

grid = GridSearchCV(
    pipeline,
    param_grid=param_grid,
    scoring="neg_root_mean_squared_error",
    cv=kf,
    verbose=2,
    n_jobs=1
)

print("\nRunning GridSearchCV...")
grid.fit(X_train, y_train)

print("\nBest params:", grid.best_params_)
print("Best CV score:", grid.best_score_)

best_model = grid.best_estimator_

# Final evaluation
y_pred_best_log = best_model.predict(X_test)
print("\nFinal RMSE (log):", np.sqrt(mean_squared_error(y_test, y_pred_best_log)))
print("Final RMSE (original):", np.sqrt(mean_squared_error(y_test_orig, np.expm1(y_pred_best_log))))

# --------------------------------------------------
# 13. SAVE MODEL
# --------------------------------------------------
joblib.dump(best_model, "houseprices_xgb_pipeline.joblib")
print("Saved model: houseprices_xgb_pipeline.joblib")

# --------------------------------------------------
# 14. SINGLE SAMPLE PREDICTION
# --------------------------------------------------
sample = X_test.iloc[[0]]
print("\nSample row:")
display(sample)

pred_log = best_model.predict(sample)[0]
pred_price = np.expm1(pred_log)

print(f"\nPredicted SalePrice (log): {pred_log:.4f}")
print(f"Predicted SalePrice: ${pred_price:,.2f}")
```
