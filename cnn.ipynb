```python
import tensorflow as tf
import numpy as np
import matplotlib.pyplot as plt
from sklearn.metrics import confusion_matrix
import seaborn as sns
from tensorflow.keras.datasets import cifar10
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import Conv2D, MaxPooling2D, Flatten, Dense

# Load the CIFAR-10 dataset
(x_train, y_train), (x_test, y_test) = cifar10.load_data()

print("Training data shape:", x_train.shape)
print("Training labels shape:", y_train.shape)
print("Testing data shape:", x_test.shape)
print("Testing labels shape:", y_test.shape)


# Normalize pixel values to the range [0, 1] by dividing by 255.0
x_train = x_train.astype('float32') / 255.0
x_test = x_test.astype('float32') / 255.0

print("Normalized training data shape:", x_train.shape)
print("Normalized testing data shape:", x_test.shape)

# Define the model architecture
model = Sequential()

# Add convolutional layer with 32 filters, a 3x3 kernel, and ReLU activation
# The input shape matches the preprocessed CIFAR-10 image data (32x32 color)
model.add(Conv2D(32, kernel_size=(3, 3), activation='relu', input_shape=(32, 32, 3)))

# Add a max pooling layer with a 2x2 pool size
model.add(MaxPooling2D(pool_size=(2, 2)))

# Add another convolutional layer with 64 filters and a 3x3 kernel
model.add(Conv2D(64, kernel_size=(3, 3), activation='relu'))

# Add another max pooling layer with a 2x2 pool size
model.add(MaxPooling2D(pool_size=(2, 2)))

# Flatten the output from the convolutional layers
model.add(Flatten())

# Add a dense layer with 128 units and ReLU activation
model.add(Dense(128, activation='relu'))

# Add the output layer with 10 units (for 10 classes in CIFAR-10) and softmax activation
model.add(Dense(10, activation='softmax'))

# Print the model summary to see the architecture
model.summary()

# Compile the model
model.compile(optimizer='adam',
              loss='sparse_categorical_crossentropy',
              metrics=['accuracy'])

# Train the model
history = model.fit(x_train, y_train, epochs=10, validation_data=(x_test, y_test))

# Get predictions for a few test images
predictions = model.predict(x_test[:10])

# CIFAR-10 class names
class_names = ['airplane', 'automobile', 'bird', 'cat', 'deer',
               'dog', 'frog', 'horse', 'ship', 'truck']

# Display the images and predictions with probability
plt.figure(figsize=(10, 10))
for i in range(10):
    plt.subplot(5, 2, i + 1)
    plt.imshow(x_test[i]) # CIFAR-10 images are already in the correct shape for imshow
    predicted_label_index = np.argmax(predictions[i])
    predicted_probability = predictions[i][predicted_label_index] * 100
    true_label_index = y_test[i][0] # y_test is a 2D array for CIFAR-10
    plt.title(f"Predicted: {class_names[predicted_label_index]} ({predicted_probability:.2f}%), True: {class_names[true_label_index]}")
    plt.axis('off')

plt.tight_layout()
plt.show()

# Evaluate the model on the test data
loss, accuracy = model.evaluate(x_test, y_test, verbose=0)

print(f"Test loss: {loss:.4f}")
print(f"Test accuracy: {accuracy:.4f}")

# Get the true labels for the test set
true_labels = y_test.flatten() # Flatten the 2D array for CIFAR-10 labels

# Get the predicted labels from the model
predicted_labels = np.argmax(model.predict(x_test), axis=-1)

# Generate the confusion matrix
cm = confusion_matrix(true_labels, predicted_labels)

# Display the confusion matrix using a heatmap
plt.figure(figsize=(10, 8))
sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=class_names, yticklabels=class_names)
plt.xlabel('Predicted Label')
plt.ylabel('True Label')
plt.title('Confusion Matrix for CIFAR-10 CNN Model')
plt.show()

```
