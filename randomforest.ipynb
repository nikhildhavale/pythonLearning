```python
# Random Forest  Classification (Breast Cancer) & Regression (Diabetes)
# Run in Google Colab / Jupyter. Requires scikit-learn, pandas, numpy, matplotlib.

# 1) IMPORTS
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay,
    mean_absolute_error, mean_squared_error, r2_score
)

plt.rcParams["figure.figsize"] = (8, 5)
np.random.seed(42)


# -----------------------------
# PART A  Random Forest CLASSIFIER (Breast Cancer)
# -----------------------------
print("\n--- Random Forest Classifier on Breast Cancer dataset ---\n")

# 2) LOAD DATA
bc = load_breast_cancer()
X_bc, y_bc = bc.data, bc.target
print("Breast cancer features shape:", X_bc.shape)
print("Target names:", bc.target_names)

# Quick DataFrame view
df_bc = pd.DataFrame(X_bc, columns=bc.feature_names)
df_bc["target"] = y_bc
print("\nClass distribution:")
print(df_bc["target"].value_counts())

# 3) TRAIN-TEST SPLIT
Xb_train, Xb_test, yb_train, yb_test = train_test_split(
    X_bc, y_bc, test_size=0.2, random_state=42)
print("\nShapes after split:", Xb_train.shape, Xb_test.shape)

# 4) TRAIN RandomForestClassifier
# Using oob_score=True gives an out-of-bag estimate (only works when bootstrap=True)
rf_clf = RandomForestClassifier(
    n_estimators=100, # 100    92, 92.5, 93
    max_depth=None, # None
    min_samples_leaf=5, # 3
    bootstrap=True,
    oob_score=True,
    n_jobs=-1,
    random_state=42
)
rf_clf.fit(Xb_train, yb_train)
print("Trained RandomForestClassifier.")
print("OOB score (estimate):", getattr(rf_clf, "oob_score_", None))

# 5) PREDICTION & EVALUATION
y_train_pred = rf_clf.predict(Xb_train)
y_test_pred  = rf_clf.predict(Xb_test)

train_acc = accuracy_score(yb_train, y_train_pred)
test_acc  = accuracy_score(yb_test, y_test_pred)
print(f"\nAccuracy -> Train: {train_acc:.4f}  Test: {test_acc:.4f}")

print("\nClassification report (Test):")
print(classification_report(yb_test, y_test_pred, target_names=bc.target_names))

cm = confusion_matrix(yb_test, y_test_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=bc.target_names)
fig, ax = plt.subplots(figsize=(5,4))
disp.plot(ax=ax, cmap="Blues", values_format='d')
plt.title("Random Forest  Confusion Matrix (Test)")
plt.show()

# 6) FEATURE IMPORTANCE
feat_imp = pd.DataFrame({
    "feature": bc.feature_names,
    "importance": rf_clf.feature_importances_
}).sort_values("importance", ascending=False).reset_index(drop=True)

print("\nTop features by importance (classifier):")
display(feat_imp.head(10))

ax = feat_imp.head(10).plot.barh(x="feature", y="importance", legend=False)
ax.invert_yaxis()
plt.title("Top 10 Feature Importances (Random Forest Classifier)")
plt.xlabel("Importance")
plt.show()

# 7) OPTIONAL: quick partial dependence / single tree extraction
# You can inspect one tree from the forest:
from sklearn import tree
one_tree = rf_clf.estimators_[3]
print("\nOne estimator depth:", one_tree.tree_.max_depth)


# -----------------------------
# PART B  Random Forest REGRESSOR (Diabetes)
# -----------------------------
print("\n\n--- Random Forest Regressor on Diabetes dataset ---\n")

# 1) LOAD DATA
db = load_diabetes()
X_db, y_db = db.data, db.target.astype(float)
print("Diabetes features shape:", X_db.shape)

df_db = pd.DataFrame(X_db, columns=db.feature_names)
df_db["target"] = y_db
print("\nFirst five rows:")
display(df_db.head(5))

# 2) TRAIN-TEST SPLIT
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_db, y_db, test_size=0.2, random_state=42)
print("\nShapes after split:", Xr_train.shape, Xr_test.shape)

# 3) TRAIN RandomForestRegressor
rf_reg = RandomForestRegressor(
    n_estimators=300,
    max_depth=None,
    min_samples_leaf=5,
    bootstrap=True,
    n_jobs=-1,
    random_state=42,
    oob_score=True
)
rf_reg.fit(Xr_train, yr_train)
print("Trained RandomForestRegressor.")
print("OOB score (regressor, R^2 estimate):", getattr(rf_reg, "oob_score_", None))

# 4) PREDICT & EVALUATE
y_pred_train = rf_reg.predict(Xr_train)
y_pred_test  = rf_reg.predict(Xr_test)

def regression_report(y_true, y_pred, label=""):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f"{label}MAE : {mae:.3f}")
    print(f"{label}MSE : {mse:.3f}")
    print(f"{label}RMSE: {rmse:.3f}")
    print(f"{label}R^2 : {r2:.4f}")

print("\nRegression performance (Train):")
regression_report(yr_train, y_pred_train, label="Train ")
print("\nRegression performance (Test):")
regression_report(yr_test, y_pred_test, label="Test  ")

# 5) DIAGNOSTIC PLOTS
plt.figure()
plt.scatter(yr_test, y_pred_test, alpha=0.7)
min_val = min(yr_test.min(), y_pred_test.min())
max_val = max(yr_test.max(), y_pred_test.max())
plt.plot([min_val, max_val], [min_val, max_val], linestyle="--")
plt.xlabel("Actual")
plt.ylabel("Predicted")
plt.title("Random Forest Regressor  Predicted vs Actual (Test)")
plt.grid(True)
plt.show()

residuals = yr_test - y_pred_test
plt.figure()
plt.scatter(y_pred_test, residuals, alpha=0.7)
plt.axhline(0.0, linestyle="--")
plt.xlabel("Predicted")
plt.ylabel("Residual (Actual - Predicted)")
plt.title("Random Forest Regressor  Residuals (Test)")
plt.grid(True)
plt.show()

# 6) FEATURE IMPORTANCES (regression)
imp_reg = pd.DataFrame({
    "feature": db.feature_names,
    "importance": rf_reg.feature_importances_
}).sort_values("importance", ascending=False).reset_index(drop=True)

print("\nTop features by importance (regressor):")
display(imp_reg.head(10))

ax = imp_reg.head(10).plot.barh(x="feature", y="importance", legend=False)
ax.invert_yaxis()
plt.title("Top 10 Feature Importances (Random Forest Regressor)")
plt.xlabel("Importance")
plt.show()

print("\nDone.")
#Random Forest + Grid Search + Random Search:
# Full updated notebook: RandomForest visualization + GridSearchCV + RandomizedSearchCV
# Works with scikit-learn, pandas, numpy, matplotlib.
# Paste whole cell into Colab / Jupyter.

import time
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.datasets import load_breast_cancer, load_diabetes
from sklearn.model_selection import train_test_split, GridSearchCV, RandomizedSearchCV
from sklearn.ensemble import RandomForestClassifier, RandomForestRegressor
from sklearn.metrics import (
    accuracy_score, classification_report, confusion_matrix, ConfusionMatrixDisplay,
    mean_absolute_error, mean_squared_error, r2_score
)
from sklearn.tree import plot_tree, export_text

plt.rcParams["figure.figsize"] = (9, 6)
np.random.seed(42)


def regression_report(y_true, y_pred, label=""):
    mae = mean_absolute_error(y_true, y_pred)
    mse = mean_squared_error(y_true, y_pred)
    rmse = np.sqrt(mse)
    r2 = r2_score(y_true, y_pred)
    print(f"{label}MAE : {mae:.3f}")
    print(f"{label}MSE : {mse:.3f}")
    print(f"{label}RMSE: {rmse:.3f}")
    print(f"{label}R^2 : {r2:.4f}")


# -------------------------
# PART A: Random Forest CLASSIFIER  Baseline, single-tree view, Grid & Randomized Search
# -------------------------
print("\n=== Random Forest CLASSIFIER: Breast Cancer ===\n")
bc = load_breast_cancer()
X_bc, y_bc = bc.data, bc.target

# Train-test split
Xb_train, Xb_test, yb_train, yb_test = train_test_split(
    X_bc, y_bc, test_size=0.2, random_state=42, stratify=y_bc
)
print("Dataset shapes (classifier) ->", Xb_train.shape, Xb_test.shape)

# Baseline RandomForestClassifier
rf_clf_base = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf_clf_base.fit(Xb_train, yb_train)
print("Baseline RandomForestClassifier trained.")

# Evaluate baseline
y_pred = rf_clf_base.predict(Xb_test)
baseline_acc = accuracy_score(yb_test, y_pred)
print(f"Baseline test accuracy: {baseline_acc:.4f}")

# -------------------------
# View a SINGLE tree from the forest (classifier)
# -------------------------
tree_idx = 0  # pick an estimator index (0 .. n_estimators-1)
single_tree = rf_clf_base.estimators_[tree_idx]
print(f"\n--- Inspecting estimator #{tree_idx} from RandomForestClassifier ---")
print("Tree depth:", single_tree.tree_.max_depth)

# Text rules (truncated)
rules_text = export_text(single_tree, feature_names=list(bc.feature_names))
print("\nFirst 600 chars of the tree rules (text):\n")
print(rules_text[:600] + ("\n... (truncated) ..." if len(rules_text) > 600 else ""))

# Plot the chosen tree (limit shown depth to keep it readable)
fig, ax = plt.subplots(figsize=(16, 6))
plot_tree(single_tree, feature_names=bc.feature_names, class_names=bc.target_names, filled=True, max_depth=3, ax=ax)
plt.title(f"RandomForestClassifier  estimator #{tree_idx} (first 3 levels shown)")
plt.show()


# -------------------------
# GRID SEARCH (CLASSIFIER)
# -------------------------
print("\n--- GridSearchCV for RandomForestClassifier (small grid) ---")
param_grid_clf = {
    "n_estimators": [100, 200],
    "max_depth": [None, 6, 12],
    "min_samples_leaf": [1, 3, 5],
    "max_features": ["sqrt", "log2"]
}

grid_clf = GridSearchCV(
    RandomForestClassifier(random_state=42, n_jobs=-1),
    param_grid=param_grid_clf,
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    verbose=1
)

t0 = time.time()
grid_clf.fit(Xb_train, yb_train)
t1 = time.time()
print(f"GridSearch finished in {t1-t0:.1f}s")
print("Best params (classifier):", grid_clf.best_params_)
print("Best CV score (classifier):", f"{grid_clf.best_score_:.4f}")

# Use best estimator
best_rf_clf = grid_clf.best_estimator_
y_test_pred = best_rf_clf.predict(Xb_test)
print(f"Test accuracy (best clf): {accuracy_score(yb_test, y_test_pred):.4f}")
print("\nClassification report (best clf):")
print(classification_report(yb_test, y_test_pred, target_names=bc.target_names))

# Confusion matrix (best clf)
cm = confusion_matrix(yb_test, y_test_pred)
disp = ConfusionMatrixDisplay(confusion_matrix=cm, display_labels=bc.target_names)
fig, ax = plt.subplots(figsize=(5, 4))
disp.plot(ax=ax, cmap="Blues", values_format='d')
plt.title("GridSearch Best RandomForestClassifier  Confusion Matrix (Test)")
plt.show()


# -------------------------
# RANDOMIZED SEARCH (CLASSIFIER)
# -------------------------
print("\n--- RandomizedSearchCV for RandomForestClassifier ---")
param_dist_clf = {
    "n_estimators": [50, 100, 200, 300],
    "max_depth": [None, 5, 8, 12, 20],
    "min_samples_leaf": [1, 2, 3, 5, 10],
    "max_features": ["sqrt", "log2", None]
}

rand_clf = RandomizedSearchCV(
    RandomForestClassifier(random_state=42, n_jobs=-1),
    param_distributions=param_dist_clf,
    n_iter=20,            # number of random combos to try
    cv=5,
    scoring="accuracy",
    n_jobs=-1,
    random_state=42,
    verbose=1
)

t0 = time.time()
rand_clf.fit(Xb_train, yb_train)
t1 = time.time()
print(f"RandomizedSearch finished in {t1-t0:.1f}s")
print("Best params (rand clf):", rand_clf.best_params_)
print("Best CV score (rand clf):", f"{rand_clf.best_score_:.4f}")

best_rand_rf_clf = rand_clf.best_estimator_
y_test_pred_rand = best_rand_rf_clf.predict(Xb_test)
print(f"Test accuracy (best rand clf): {accuracy_score(yb_test, y_test_pred_rand):.4f}")


# Inspect one tree from the tuned grid-search forest (classifier)
print("\n--- Inspecting one estimator from Grid-searched RandomForestClassifier ---")
tuned_tree = best_rf_clf.estimators_[0]
print("Tuned tree depth:", tuned_tree.tree_.max_depth)
fig, ax = plt.subplots(figsize=(14, 6))
plot_tree(tuned_tree, feature_names=bc.feature_names, class_names=bc.target_names, filled=True, max_depth=3, ax=ax)
plt.title("One tree from Grid-searched RandomForestClassifier (first 3 levels shown)")
plt.show()


# -------------------------
# PART B: Random Forest REGRESSOR  Baseline, single-tree view, Grid & Randomized Search
# -------------------------
print("\n\n=== Random Forest REGRESSOR: Diabetes ===\n")
db = load_diabetes()
X_db, y_db = db.data, db.target.astype(float)

# Train-test split (regression)
Xr_train, Xr_test, yr_train, yr_test = train_test_split(X_db, y_db, test_size=0.2, random_state=42)
print("Dataset shapes (regressor) ->", Xr_train.shape, Xr_test.shape)

# Baseline RandomForestRegressor
rf_reg_base = RandomForestRegressor(n_estimators=100, random_state=42, n_jobs=-1)
rf_reg_base.fit(Xr_train, yr_train)
print("Baseline RandomForestRegressor trained.")

# Baseline evaluation
y_pred_reg = rf_reg_base.predict(Xr_test)
print("\nBaseline regressor test performance:")
regression_report(yr_test, y_pred_reg, label="Baseline Test ")

# -------------------------
# View a SINGLE tree from the regressor forest
# -------------------------
tree_idx = 0
single_reg_tree = rf_reg_base.estimators_[tree_idx]
print(f"\n--- Inspecting estimator #{tree_idx} from RandomForestRegressor ---")
print("Tree depth:", single_reg_tree.tree_.max_depth)

rules_text_reg = export_text(single_reg_tree, feature_names=list(db.feature_names))
print("\nFirst 500 chars of rules (text):\n")
print(rules_text_reg[:500] + ("\n... (truncated) ..." if len(rules_text_reg) > 500 else ""))

fig, ax = plt.subplots(figsize=(14, 6))
plot_tree(single_reg_tree, feature_names=db.feature_names, filled=True, max_depth=3, ax=ax)
plt.title(f"RandomForestRegressor  estimator #{tree_idx} (first 3 levels shown)")
plt.show()


# -------------------------
# GRID SEARCH (REGRESSOR)
# -------------------------
print("\n--- GridSearchCV for RandomForestRegressor (small grid) ---")
param_grid_reg = {
    "n_estimators": [100, 200],
    "max_depth": [None, 6, 12],
    "min_samples_leaf": [1, 3, 5],

}

grid_reg = GridSearchCV(
    RandomForestRegressor(random_state=42, n_jobs=-1),
    param_grid=param_grid_reg,
    cv=5,
    scoring="r2",
    n_jobs=-1,
    verbose=1
)

t0 = time.time()
grid_reg.fit(Xr_train, yr_train)
t1 = time.time()
print(f"GridSearch (reg) finished in {t1-t0:.1f}s")
print("Best params (reg):", grid_reg.best_params_)
print("Best CV R^2 (reg):", f"{grid_reg.best_score_:.4f}")

best_rf_reg = grid_reg.best_estimator_
y_test_pred_reg = best_rf_reg.predict(Xr_test)
print("\nTest performance (best grid reg):")
regression_report(yr_test, y_test_pred_reg, label="Test ")


# -------------------------
# RANDOMIZED SEARCH (REGRESSOR)
# -------------------------
print("\n--- RandomizedSearchCV for RandomForestRegressor ---")
param_dist_reg = {
    "n_estimators": [50, 100, 200, 300],
    "max_depth": [None, 5, 8, 12, 20],
    "min_samples_leaf": [1, 2, 3, 5, 10],
    "max_features": ["auto", "sqrt", "log2"]
}

rand_reg = RandomizedSearchCV(
    RandomForestRegressor(random_state=42, n_jobs=-1),
    param_distributions=param_dist_reg,
    n_iter=20,
    cv=5,
    scoring="r2",
    n_jobs=-1,
    random_state=42,
    verbose=1
)

t0 = time.time()
rand_reg.fit(Xr_train, yr_train)
t1 = time.time()
print(f"RandomizedSearch (reg) finished in {t1-t0:.1f}s")
print("Best params (rand reg):", rand_reg.best_params_)
print("Best CV R^2 (rand reg):", f"{rand_reg.best_score_:.4f}")

best_rand_rf_reg = rand_reg.best_estimator_
y_test_pred_rand_reg = best_rand_rf_reg.predict(Xr_test)
print("\nTest performance (best rand reg):")
regression_report(yr_test, y_test_pred_rand_reg, label="Test ")


# Inspect one estimator from tuned regressor
print("\n--- Inspecting one estimator from Grid-searched RandomForestRegressor ---")
tuned_reg_tree = best_rf_reg.estimators_[0]
print("Tuned reg tree depth:", tuned_reg_tree.tree_.max_depth)
fig, ax = plt.subplots(figsize=(14, 6))
plot_tree(tuned_reg_tree, feature_names=db.feature_names, filled=True, max_depth=3, ax=ax)
plt.title("One tree from Grid-searched RandomForestRegressor (first 3 levels shown)")
plt.show()

print("\nAll searches complete.")
```
